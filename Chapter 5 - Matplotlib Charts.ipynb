### Creating beautiful chars using Matplotlib

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.gca(projection = '3d')
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
surf = ax.plot_surface(X, Y, Z, rstride = 1, cstride = 1, cmap = cm.coolwarm, linewidth = 0, antialiased = False)

ax.set_zlim(-1.01, 1.01)

ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

fig.colorbar(surf, shrink = 0.5, aspect = 5)

plt.show()
```

### Creating our first chart

```python
import numpy as np
import matplotlib.pyplot as plot
from pylab import show

x = np.arange(0, 5, 0.1)
y = np.sin(x)
plt.plot(x, y)

show()
```

```python
from pylab import show, arange, sin, plot, pi

t = arange(0.0, 2.0, 0.01)
s = sin(2*pi*t)
plot(t, s)

show()
```

### Placing labels on charts

```python
from matplotlib.figure import Figure
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import tkinter as tk

fig = Figure(figsize = (12, 8), facecolor = 'white')

#axis = fig.add_subplot(111)
axis = fig.add_subplot(211)
axis1 = fig.add_subplot(212)

xValues = [1, 2, 3, 4]
yValues = [5, 7, 6, 8]

xValues1 = [1, 2, 3, 4]
yValues1 = [7, 5, 8, 6]

axis.plot(xValues, yValues)
axis1.plot(xValues1, yValues1)

axis.set_xlabel('Horizontal Label')
axis.set_ylabel('Vertical Label')

axis.grid(linestyle = '-')    #solid grid lines
axis1.grid()

def _destroyWindow():
    root.quit()
    root.destroy()
    
root = tk.Tk()
root.withdraw()
root.protocol('WM_DELETE_WINDOW', _destroyWindow)

canvas = FigureCanvasTkAgg(fig, master = root)
canvas._tkcanvas.pack(side = tk.TOP, fill = tk.BOTH, expand = 1)

root.update()
root.deiconify()
root.mainloop()
```

### How to give the chart a legend

```python
from matplotlib.figure import Figure
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import tkinter as tk

fig = Figure(figsize = (12, 8), facecolor = 'white')

#axis = fig.add_subplot(111)
axis = fig.add_subplot(211)

xValues = [1, 2, 3, 4]

yValues0 = [6, 7.5, 8, 7.5]
yValues1 = [5.5, 6.5, 8, 6]
yValues2 = [6.5, 7, 8, 7]

t0, = axis.plot(xValues, yValues0)
t1, = axis.plot(xValues, yValues1)
t2, = axis.plot(xValues, yValues2)

axis.set_xlabel('Horizontal Label')
axis.set_ylabel('Vertical Label')

axis.grid(linestyle = '-')    #solid grid lines

fig.legend((t0, t1, t2), ('First line', 'Second line', 'Third line'), 'upper right')

def _destroyWindow():
    root.quit()
    root.destroy()
    
root = tk.Tk()
root.withdraw()
root.protocol('WM_DELETE_WINDOW', _destroyWindow)

canvas = FigureCanvasTkAgg(fig, master = root)
canvas._tkcanvas.pack(side = tk.TOP, fill = tk.BOTH, expand = 1)

root.update()
root.deiconify()
root.mainloop()
```
